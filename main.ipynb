```python
import pandas as pd 
import duckdb
df = pd.read_csv("financials.csv.xls", sep=",", engine="python")

df.to_csv("clean_data.csv", index=False)
```

```python
# connect to duckdb file 
con = duckdb.connect("securities.duckdb")

# create table from dataframe 
con.register("temp_df", df)
con.execute("""
            CREATE OR REPLACE TABLE stocks AS 
            SELECT * FROM temp_df 
            """)
print(con.execute("SELECT * FROM stocks LIMIT 5").fetchdf()) 

# close connection 
con.close()
```
